crea una funcion que reciba 2 numeros y una operacion ( suma, resta, multiplicacion, division) y devuelva el resultado

```python
def mi_calculo(num1,num2,operacion):
    if operacion == "suma":
        return num1 + num2
    elif operacion == "resta":
        return num1 - num2
    elif operacion =="multiplicacion":
        return num1 * num2
    else:
        return num1/num2
    
```

```python
mi_calculo(10,5,"suma")
```

conversor de moneda crea una funcion que convierta de pesos a dolares, que reciba como parametro la cantidad de pesos y el valor de dolar y devuelva la cantidad de dolares.

```python
def conversor_moneda(pesos,cotizacion):
    dolares=pesos/cotizacion
    return dolares
```

```python
conversor_moneda(1000000,1205)
```
